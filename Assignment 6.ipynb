Define the following Terms :
1. Class
2. Object
3. Attribute
4. Behavior

Ans:
1) Class: Class is the blueprint of the Object. All the properties and behaviours of Object are defined inside a class as attributes and methods.

2) Object: Object is the real world model of class or running form of class. Object contains all the properties and behaviors defined inside class. Object is used to access the data of class.

3) Attribute: Attributes are the variables that are defined inside a class.

Question 5.

Write a code in Python in which create a class named Car which have 5 Attributes such like (model, color and name etc.) and 3 methods. And create 5 Object instance from that class.


```python
#Question 5.

#Write a code in Python in which create a class named Car which have
#5 Attributes such like (model, color and name etc.) and 3 methods. And create 5 Object instance from that class


class Car():
    # Init method
    def __init__(self, name, model, color, Drivetype, Horsepower):
        self.name = name
        self.model = model
        self.color = color
        self.Drivetype = Drivetype
        self.Horsepower = Horsepower
        
    # show_properties method
    def show_properties(self):
        print("Car Properties")
        print(f"Name         : {self.name}")
        print(f"Model        : {self.model}")
        print(f"Color        : {self.color}")
        print(f"Drivetype      : {self.Drivetype}")
        print(f"Horsepower : {self.Horsepower} \n")
        
    
    # update_name method
    def update_name(self, newname):
        self.name = newname
    
    # update_color method
    def update_color(self, newcolor):
        self.color = newcolor


        
# Objects

car1 = Car("Rolls-Royce's", 2020, "Silver Shadow", "Rear-wheel", "6749 CC")
car2 = Car("Ferrari", 2019, "Brown", "Four-wheel", "6262 CC")
car3 = Car("lamborghini", 2018, "Black Shine", "Rear-wheel", "6498 CC")
car4 = Car("Range-Rover", 2017, "Grey", "Four-wheel", "7798 CC")
car5 = Car("Civic", 2020, "Violet", "Rear-wheel", "2000 CC")

# Calling methods
car1.show_properties()
car2.show_properties()
car3.show_properties()
car4.show_properties()
car5.show_properties()
        
    
# Changing name
 # before Changing
print("Before Changing : ",car1.name)
car1.update_name("Audi")
print("After Changing : ",car1.name)
```
